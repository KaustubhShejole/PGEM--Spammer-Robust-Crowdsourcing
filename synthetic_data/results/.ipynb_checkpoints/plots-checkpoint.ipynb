```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure N and K are treated as integers for proper sorting/labelling
df['N'] = df['N'].astype(int)
df['K'] = df['K'].astype(int)

# 1. Create a "Setting" column for the X-axis label
# This combines N and K into a single string like "(N=10, K=20)"
df['Setting'] = df.apply(lambda row: f"(N={row['N']}, K={row['K']})", axis=1)

# Sort the dataframe to ensure the settings appear sequentially on the X-axis
# Adjust the sorting list if you want a different order (e.g., sort by K then N)
df = df.sort_values(by=['N', 'K'])

# Extract the sorted settings and create an index for plotting
settings_order = df['Setting'].values
x_indices = np.arange(len(settings_order))

# Define the parameters to loop through
methods = ['PGEM', 'BT', 'FactorBT', 'CrowdBT']
metrics = ['acc', 'wacc', 'tau']
colors = sns.color_palette("husl", len(methods)) # High-contrast colors
markers = ['o', 's', '^', 'D'] # distinct markers for black & white printing compatibility

# Setup the figure (1 row, 3 columns)
plt.figure(figsize=(18, 6))
sns.set_style("ticks") # Clean white background with ticks
sns.set_context("paper", font_scale=1.5) # Larger fonts for readability in papers

for i, metric in enumerate(metrics):
    ax = plt.subplot(1, 3, i+1)
    
    for j, method in enumerate(methods):
        # Construct column names
        mean_col = f"{method}_{metric}_mean"
        std_col = f"{method}_{metric}_std"
        
        # Get data
        means = df[mean_col].values
        stds = df[std_col].values
        
        # Plot line with error bars
        ax.errorbar(x_indices, means, yerr=stds, label=method, 
                    marker=markers[j], markersize=8, 
                    capsize=5, elinewidth=1.5, linewidth=2, 
                    color=colors[j])
        
    # Formatting the subplot
    ax.set_title(metric.upper(), fontsize=16, fontweight='bold')
    ax.set_xticks(x_indices)
    ax.set_xticklabels(settings_order, rotation=45, ha='right')
    ax.set_xlabel('Setting (N, K)', fontsize=14)
    ax.set_ylabel(metric, fontsize=14)
    ax.grid(True, linestyle='--', alpha=0.5) # Light grid for guidance
    
    # Add legend only to the first plot to avoid clutter
    if i == 0:
        ax.legend(title="Method", loc='best', frameon=True, framealpha=0.9)

plt.tight_layout()
plt.savefig('comparison_results.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure N and K are treated as integers
df['N'] = df['N'].astype(int)
df['K'] = df['K'].astype(int)

# Create "Setting" column for the X-axis label
df['Setting'] = df.apply(lambda row: f"(N={row['N']}, K={row['K']})", axis=1)

# Sort the dataframe
df = df.sort_values(by=['N', 'K'])

# Extract the sorted settings and create an index for plotting
settings_order = df['Setting'].values
x_indices = np.arange(len(settings_order))

# Define the parameters
methods = ['PGEM', 'BT', 'FactorBT', 'CrowdBT', 'BARP', 'RC']
metrics = ['acc', 'wacc', 'tau']
colors = sns.color_palette("husl", len(methods))

# --- FIX: Extended markers list to handle 6 methods ---
markers = ['o', 's', '^', 'D', 'v', 'X'] 

# Styling
sns.set_style("ticks")
sns.set_context("paper", font_scale=1.5)

for metric in metrics:
    plt.figure(figsize=(10, 6)) 
    ax = plt.gca()
    
    for j, method in enumerate(methods):
        mean_col = f"{method}_{metric}_mean"
        std_col = f"{method}_{metric}_std"
        
        means = df[mean_col].values
        stds = df[std_col].values
        
        ax.errorbar(x_indices, means, yerr=stds, label=method, 
                    marker=markers[j], markersize=10, 
                    capsize=6, elinewidth=2, linewidth=2.5, 
                    color=colors[j])
        
    # Formatting
    title_map = {'acc': 'Accuracy', 'wacc': 'Weighted Accuracy', 'tau': "Kendall's Tau"}
    plot_title = title_map.get(metric, metric.upper())
    
    ax.set_title(plot_title, fontsize=18, fontweight='bold')
    ax.set_xticks(x_indices)
    ax.set_xticklabels(settings_order, rotation=45, ha='right')
    ax.set_xlabel('Setting (N, K)', fontsize=16)
    ax.set_ylabel(metric, fontsize=16)
    ax.grid(True, linestyle='--', alpha=0.5)
    
    # Updated legend to handle more items gracefully
    ax.legend(title="Method", loc='best', fontsize=12, title_fontsize=13, 
              frameon=True, framealpha=0.9, ncol=2)

    plt.tight_layout()
    plt.savefig(f'comparison_{metric}.png', dpi=300, bbox_inches='tight')
    plt.show()
```
