```python
import pandas as pd

# Read CSV (adjust separator if needed)
df = pd.read_csv("results/0_1_comparison_metrics.csv")

def fmt(mean, std=None):
    if std is None:
        return f"{mean:.3f}"
    return f"{mean:.3f} $\\pm$ {std:.3f}"

rows = []

for _, r in df.iterrows():
    row = []
#     print(r)

    # K and M (your table uses K, M instead of N, K)
    row.append(str(int(r["K"])))
    row.append(str(int(r["N"])))  # assuming M = N

    # ---- Ours (PGEM) ----
    row += [
        fmt(r["PGEM_acc_mean"], r["PGEM_acc_std"]),
        fmt(r["PGEM_wacc_mean"], r["PGEM_wacc_std"]),
        fmt(r["PGEM_tau_mean"], r["PGEM_tau_std"]),
    ]

    # ---- FactorBT ----
    row += [
        fmt(r["FactorBT_acc_mean"], r["FactorBT_acc_std"]),
        fmt(r["FactorBT_wacc_mean"], r["FactorBT_wacc_std"]),
#         fmt(r["FactorBT_tau_mean"], r["FactorBT_tau_std"]),
    ]

    # ---- Std-BTL (BT) ----
    row += [
        fmt(r["BT_acc"]),
        fmt(r["BT_wacc"]),
        fmt(r["BT_tau"]),
    ]

    # ---- CrowdBT ----
    row += [
        fmt(r["CrowdBT_acc_mean"], r["CrowdBT_acc_std"]),
        fmt(r["CrowdBT_wacc_mean"], r["CrowdBT_wacc_std"]),
        fmt(r["CrowdBT_tau_mean"], r["CrowdBT_tau_std"]),
    ]

    # ---- RC ----
    row += [
        fmt(r["RC_acc"]),
        fmt(r["RC_wacc"]),
        fmt(r["RC_tau"]),
    ]

    # ---- BARP ----
    row += [
        fmt(r["BARP_acc"]),
        fmt(r["BARP_wacc"]),
        fmt(r["BARP_tau"]),
    ]

    rows.append(" & ".join(row) + r" \\")

# Print LaTeX rows
for line in rows:
    print(line)
```
